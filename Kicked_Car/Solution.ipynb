```python
missing_PRIMEUNIT = np.sum(pd.isna(train.PRIMEUNIT))
missing_AUCGUART = np.sum(pd.isna(train.AUCGUART))

print("Ratio of missing data for PRIMEUNIT is: ", 
     np.round(missing_PRIMEUNIT/train.shape[0], 2))
print("Ratio of missing data for AUCGUART is: ", 
     np.round(missing_AUCGUART/train.shape[0], 2))
```

```python
train[['PRIMEUNIT', 'IsBadBuy']].groupby(['PRIMEUNIT'], as_index = False).mean()
```

```python
train[['PRIMEUNIT']] = train[['PRIMEUNIT']].fillna('MAYBE')
train[['PRIMEUNIT', 'IsBadBuy']].groupby(['PRIMEUNIT'], as_index = False).mean()
```

```python
train['AUCGUART'] = train[['AUCGUART']].fillna('MAYBE')
train[['AUCGUART', 'IsBadBuy']].groupby(['AUCGUART'], as_index = False).mean()
```

**Since there's a large difference with the missing values and the others, we'll leave the values with a placeholder.**

**There are still some values that are missing. If the testing set has missing values as well then we will fill the values with a guess at the missing data.**

```python
print("Shape before dropping missing values",test.shape)
print("Shape after dropping missing values", test.dropna().shape)
```

**Appears that the test set probably has the same missing data in the columns as the training set.**

```python
missing_PRIMEUNIT = np.sum(pd.isna(test.PRIMEUNIT))
missing_AUCGUART = np.sum(pd.isna(test.AUCGUART))

print("Ratio of missing data for PRIMEUNIT is: ", 
     np.round(missing_PRIMEUNIT/test.shape[0], 2))
print("Ratio of missing data for AUCGUART is: ", 
     np.round(missing_AUCGUART/test.shape[0], 2))
```

**A quick verification of the data shows that the test set has a similar profile missing data primarily in the PRIMEUNIT and AUCGUART sections. We'll apply the same preprocessing step as before.**

```python
test[['PRIMEUNIT']] = test[['PRIMEUNIT']].fillna('MAYBE')
test['AUCGUART'] = test[['AUCGUART']].fillna('MAYBE')
```

```python
print("Shape before dropping missing values",test.shape)
print("Shape after dropping missing values", test.dropna().shape)
```

**Since some data appears to be missing for the testing set, we will fill in all missing categorical data with a filler category of 'MAYBE' and all missing numerical data with the respective average value. This will be done for both training and testing set.**

```python
for i in train:
    if train[i].isna().any():
        print(i, 'has missing data')
        if train[i].dtypes == 'object':
            print('Filling with filler value for', i)
            train[i] = train[i].fillna('MAYBE')
        else:
            print('filling new continuous value')
            train[i] = train[i].fillna(train[i].mean())
```

```python
for i in test:
    if test[i].isna().any():
        print(i, 'has missing data')
        if test[i].dtypes == 'object':
            print('Filling with filler value for', i)
            test[i] = test[i].fillna('MAYBE')
        else:
            print('filling new continuous value')
            test[i] = test[i].fillna(test[i].mean())
```

**Next we will convert text data to a numerical value. This will be only applied to any categorical data. We will exclude dates from this since vehicle age relates to the date it was purchased. We will drop RefId as well since this is not additional data about the vehicle and is rather an ID for data processing use.**

**We will convert all strings to UPPER so there's no separate categories with different spellings.**

```python
train = train.drop(['PurchDate', 'RefId'], axis = 1)
test = test.drop(['PurchDate', 'RefId'], axis = 1)
```

```python
for i in train:
    if train[i].dtypes == 'object':
        #convert all string data to upper
        train[i] = train[i].str.upper()
```

```python
for i in test:
    if test[i].dtypes == 'object':
        #convert all string data to upper
        test[i] = test[i].str.upper()
```

```python
indexing_dict = {}
for i in train:
    if train[i].dtypes == 'object':
        train[i], hold = pd.factorize(train[i])
        indexing_dict[i] = hold
        print('Number of uniques in', i, len(hold))
```

**Applying the new labels from the training factorization to testing set now.**

```python
#before
test.head()
```

```python
def convert_labels(index, value):
    if value in index:
        return index.get_loc(value)
    else:
        return -1
```

```python
for i in test:
    if test[i].dtypes == 'object':
        test[i] = test[i].apply(lambda x: convert_labels(indexing_dict[i],x))
```

```python
#after
test.head()
```

### First Model Trained (Logistic Regression)
**Let's look at an initial classifier before looking at the data**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
```

```python
X = train[train.axes[1][1:]]
y = train['IsBadBuy'];

#for immediate results, we'll split the training set further into a train/test set
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 1)
```

```python
clf = LogisticRegression().fit(X_train, y_train)
print('Training accuracy:', clf.score(X_train, y_train))
print('Testing accuracy:', clf.score(X_test, y_test))
```

**This looks really promising!!! Checking the predicted values now**

```python
y_pred = clf.predict(X_test)
```

```python
#accounting for how many positive and negative predictions were made
y_pred_dict = {}
for i in y_pred:
    if i in y_pred_dict:
        y_pred_dict[i] +=1
    else:
        y_pred_dict[i] = 1
```

```python
y_pred_dict
```

**The model we trained appears to be guessing that only 0s. We can confirm this by checking the percent negative ratio in the training and testing set.**

```python
print("Percentage of negative values in training set:", np.round(1-sum(y_train)/len(y_train), 6))
print("Percentage of negative values in testing set:", np.round(1-sum(y_test)/len(y_test), 6))
```

**The percent negative values in the training and testing set are exactly the same as the model predicted. Thus it appears the mismatch of negative to positive samples is preventing our model from making any sort of meaningful predictions. Thus we will create a new training set that will have an equal number of positive and negative examples to train.**

```python
bad_buys = train[train["IsBadBuy"] == 1]
good_buys= train[train["IsBadBuy"] == 0]
print(bad_buys.shape[0], good_buys.shape[0])
```

```python
good_buys = good_buys.sample(n = bad_buys.shape[0])
print(bad_buys.shape[0], good_buys.shape[0])
```

**With an equal number of positive and negative examples we will now create a new balanced training set**

```python
balanced_Train = pd.concat([good_buys, bad_buys])

#shuffling the order of the samples now
balanced_Train = balanced_Train.sample(frac = 1).reset_index(drop=True)
```

```python
balanced_Train.head()
```

```python
balanced_Train.tail()
```

```python
balanced_X = balanced_Train[balanced_Train.axes[1][1:]]
balanced_y = balanced_Train['IsBadBuy']
```

```python
bX_train, bX_test, by_train, by_test = train_test_split(balanced_X,
                                                        balanced_y,
                                                        test_size = 0.2,
                                                        random_state = 1)
```

**Now to see how the data will fit using logistic regression using all the provided features**

```python
clf = LogisticRegression().fit(bX_train, by_train)
print('Balanced training accuracy:', clf.score(bX_test, by_test))
print('Balanced testing accuracy:', clf.score(bX_test, by_test))
print('Overall test accuracy:', clf.score(X_test, y_test))
```

**This is a better result since the model is actually predicting something using all of the available data. Checking Kaggle score next by training classifier on all training points now and predicting the test set.**

```python
clf = LogisticRegression().fit(balanced_X, balanced_y)
print('Overall training accuracy:', clf.score(balanced_X, balanced_y))
```
